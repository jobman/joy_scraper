```python
handle = open("all_file_names.txt", "r")
file_names = []
for line in handle:
    file_names.append(line)
handle.close()
print(len(file_names))
```

```python
import numpy as np
import pandas as pd
```

```python
from datetime import datetime

dates = []
ratings = []
indexes = []
extensions = []
timestamps = []

for filename in file_names:
    timestamp, rating, index, extension = filename.split()
    
    timestamp = int(timestamp)
    dt_object = datetime.fromtimestamp(timestamp)
    date = dt_object.strftime("%d/%m/%Y")
    
    timestamps.append(int(timestamp))
    dates.append(date)
    ratings.append(float(rating))
    indexes.append(index)
    extensions.append(extension)
    
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(timestamps,ratings)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x = np.array(timestamps)
x=x[:,np.newaxis]
y = np.array(ratings)
y=y[:,np.newaxis]
lr = LinearRegression().fit(x,y)
prediction = lr.predict(x)

plt.plot(x,y, color='black')
plt.plot(x,prediction)

#Обрезаем все что ниже регрессии
new_files = []
for prediction,rating,filename in zip(lr.predict(x),ratings,file_names):
    if prediction<rating:
        new_files.append(filename)
```
