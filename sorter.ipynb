```python
#Получить список всех файлов
from os import walk

f = []
for (dirpath, dirnames, filenames) in walk("images/"):
    f.extend(filenames)
    break
```

```python
len(f)
```

```python
#Сохранить список файлов на диск
with open('all_file_names.txt', 'w') as file:
    for item in f:
        file.write("%s\n" % item)
```

```python
file_names = []
with open('best.txt', 'r') as file:
    for item in file:
        file_names.append(item)
len(file_names)
```

```python
file_names
```

```python
files=[]
for i in range(len(file_names)):
    if i%2==0:
        files.append(file_names[i])
```

```python
files_normal = []
for file in files:
    files_normal.append(file[:-1])
```

```python
files_normal
```

```python
from shutil import copyfile
for src in files_normal:
    copyfile("images/"+src, "best/"+src)
    print("Copied")
    print(src)
    print()
```
